```python
'''Write a Python Code for following Edge Detection Operators:
1. Roberts Operator
2. Prewitt’s Operator
3. Sobel Operator
4. Prewitt’s Compass Operators
5. Sobel Compass Operators'''
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('D:\TY ET A\CV\sunflower.jpg', 0)
plt.imshow(img,cmap="gray")
plt.title('Original')
plt.show()
```

```python
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
plt.imshow(img_gaussian,cmap="gray")
plt.title('gaussian blur')
plt.show()
```

```python
def robert():
    kernelx = np.array([[1,0],[0,-1]],dtype=int)
    kernely = np.array([[0,1],[-1,0]],dtype=int)
    img_robertx = cv2.filter2D(img_gaussian, -1, kernelx)
    img_roberty = cv2.filter2D(img_gaussian, -1, kernely)

    plt.imshow(img_robertx,cmap="gray")
    plt.title('Robert Operatorx')
    plt.show()
    plt.imshow(img_roberty,cmap="gray")
    plt.title('Robert Operatory')
    plt.show()

    img_robertxy=img_roberty+img_robertx
    plt.imshow(img_robertxy,cmap="gray")
    plt.title('Robert Operator')
    plt.show()
```

```python
def perwitt():
    kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]],dtype=int)
    kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]],dtype=int)
    img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
    img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)

    plt.imshow(img_prewittx,cmap="gray")
    plt.title('Perwitt Operatorx')
    plt.show()
    plt.imshow(img_prewitty,cmap="gray")
    plt.title('Perwitt Operatory')
    plt.show()

    img_prewittxy=img_prewitty+img_prewittx
    plt.imshow(img_prewittxy,cmap="gray")
    plt.title('Perwitt Operator')
    plt.show()
```

```python
def sobel():
    kernelx = np.array([[-1,-5,-1],[0,0,0],[1,5,1]],dtype=int)
    kernely = np.array([[-1,0,1],[-5,0,5],[-1,0,1]],dtype=int)
    img_sobelx = cv2.filter2D(img_gaussian, -1, kernelx)
    img_sobely = cv2.filter2D(img_gaussian, -1, kernely)

    plt.imshow(img_sobelx,cmap="gray")
    plt.title('Sobel Operatorx')
    plt.show()
    plt.imshow(img_sobely,cmap="gray")
    plt.title('Sobel Operatory')
    plt.show()

    img_sobelxy=img_sobely+img_sobelx
    plt.imshow(img_sobelxy,cmap="gray")
    plt.title('Sobel Operator')
    plt.show()
```

```python
def perwitt_compass():
    kernel0 = np.array([[-1,-1,-1],[0,0,0],[1,1,1]],dtype=int)
    kernel45 = np.array([[-1,-1,0],[-1,0,1],[0,1,1]],dtype=int)
    kernel90 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]],dtype=int)
    kernel135 = np.array([[0,1,1],[-1,0,1],[-1,-1,0]],dtype=int)
    kernel180 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]],dtype=int)
    kernel225 = np.array([[1,1,0],[1,0,-1],[0,-1,-1]],dtype=int)
    kernel270 = np.array([[1,0,-1],[1,0,-1],[1,0,-1]],dtype=int)
    kernel315 = np.array([[0,-1,-1],[1,0,-1],[1,1,0]],dtype=int)


    img_pcompass = cv2.filter2D(img_gaussian, -1,kernel0)+cv2.filter2D(img_gaussian, -1,kernel45)+cv2.filter2D(img_gaussian, -1,kernel90)+cv2.filter2D(img_gaussian, -1,kernel135)+cv2.filter2D(img_gaussian, -1,kernel180)+cv2.filter2D(img_gaussian, -1,kernel225)+cv2.filter2D(img_gaussian, -1,kernel270)+cv2.filter2D(img_gaussian, -1,kernel315)

    plt.imshow(img_pcompass,cmap="gray")
    plt.title('Perwitt compass Operator')
    plt.show()
```

```python
def sobel_compass():
    kernel0 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]],dtype=int)
    kernel45 = np.array([[-2,-1,0],[-1,0,1],[0,1,2]],dtype=int)
    kernel90 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]],dtype=int)
    kernel135 = np.array([[0,1,-2],[-1,0,1],[2,-1,0]],dtype=int)
    kernel180 = np.array([[1,2,1],[0,0,0],[-1,-2,-1]],dtype=int)
    kernel225 = np.array([[2,1,0],[1,0,-1],[0,-1,-2]],dtype=int)
    kernel270 = np.array([[1,0,-1],[2,0,-2],[1,0,-1]],dtype=int)
    kernel315 = np.array([[0,-1,-2],[1,0,-1],[2,1,0]],dtype=int)


    img_scompass = cv2.filter2D(img_gaussian, -1,kernel0)+cv2.filter2D(img_gaussian, -1,kernel45)+cv2.filter2D(img_gaussian, -1,kernel90)+cv2.filter2D(img_gaussian, -1,kernel135)+cv2.filter2D(img_gaussian, -1,kernel180)+cv2.filter2D(img_gaussian, -1,kernel225)+cv2.filter2D(img_gaussian, -1,kernel270)+cv2.filter2D(img_gaussian, -1,kernel315)

    plt.imshow(img_scompass,cmap="gray")
    plt.title('Sobel compass Operator')
    plt.show()
```

```python
x=int(input('Select the operator you want to perform on input image.\n 1: Robert 2:Perwitt 3:Sobel 4:Perwitt Compass 5:Sobel compass : '))
while(x!=0):
    if x==1:robert()
    elif x==2:perwitt()
    elif x==3:sobel()
    elif x==4: perwitt_compass()
    elif x==5: sobel_compass()
    else: print('Invalid Choice')
    x=int(input('Select the operator you want to perform on input image.\n 1:Perwitt 2:Sobel 3:Perwitt Compass 4:Sobel compass : '))
        
```
