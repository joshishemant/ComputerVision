```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('D:\TYETA\CV\shape.jpg',0)
img2=cv2.imread('D:\TYETA\CV\shape.jpg')
```

```python
gray = np.float32(img)
dst = cv2.cornerHarris(gray,2,3,0.04)
dst = cv2.dilate(dst,None)
img2[dst>0.0004*dst.max()]=[255,0,0]
plt.imshow(img2)
plt.title('Harrish Corner')
plt.show()
```
