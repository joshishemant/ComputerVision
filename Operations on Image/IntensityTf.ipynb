```python
#Write a Python Code for following Intensity Transformations:
#1. Intensity Level Slicing Without Background
#2. Intensity Level Slicing With Background
#3. Log
#4. Power Law
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('D:\TY ET A\CV\sunflower.jpg',0)
plt.imshow(img,cmap="gray")
plt.title('Original')
plt.show()
```

```python
def slice_nobg():
    T1=80
    T2=150
    m,n = img.shape
    img_thresh_back = np.zeros((m,n), dtype = int)
    for i in range(m):
        for j in range(n):
            if T1 < img[i,j] < T2:
                img_thresh_back[i,j]= 255
            else:
                img_thresh_back[i,j] = img[i,j]
    plt.imshow(img_thresh_back,cmap="gray")
    plt.title(' Level Slicing Without Background')
    plt.show()
```

```python
def slice_bg():
    T1=50
    T2=120
    m,n = img.shape
    img_thresh = np.zeros((m,n), dtype = int)
    for i in range(m):
        for j in range(n):
            if T1 < img[i,j] < T2:
                img_thresh[i,j]= 255
            else:
                img_thresh[i,j] = 0
    plt.imshow(img_thresh,cmap="gray")
    plt.title('Level Slicing With Background')
    plt.show()
```

```python
def logtrans():
    c = 255/(np.log(1 + np.max(img)))
    log_transformed = c * np.log(1 + img)
    log_transformed = np.array(log_transformed, dtype = np.uint8)
    plt.imshow(log_transformed,cmap="gray")
    plt.title('log transformed')
    plt.show()
```

```python
def powerlaw():
    gamma=float(input("Enter the value of Gamma: "))  
    gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8') 
    plt.imshow(gamma_corrected,cmap="gray")
    plt.title('gamma corrected')
    plt.show()
```

```python
x=int(input('Select the Transformation you want to perform on input image.\n 1. Intensity Level Slicing Without Background 2. Intensity Level Slicing With Background 3. Log 4. Power Law '))
while(x!=0):
    if x==1: slice_nobg()
    elif x==2: slice_bg()
    elif x==3:logtrans()
    elif x==4:powerlaw()
    else: print('Invalid Choice')
    x=int(input('Select the Transformation you want to perform on input image.\n 1. Intensity Level Slicing Without Background 2. Intensity Level Slicing With Background 3. Log 4. Power Law '))
```
