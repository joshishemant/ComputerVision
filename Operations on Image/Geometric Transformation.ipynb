```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('D:\TY ET A\CV\sunflower.jpg',0)
```

```python
plt.imshow(img,cmap="gray")
plt.title('Original')
plt.show()
```

```python
print (img.shape)
```

```python
def resize():
    r = 200.0 / img.shape[1]
    dim = (100, int(img.shape[0] * r))
    resized = cv2.resize(img, dim)
    plt.imshow(resized,cmap="gray")
    plt.title('resized')
    plt.show()
```

```python
def rotate():
    (h, w) = img.shape[:2]
    center = (w / 2, h / 2)
    angle=int(input("Enter the rotation angle in degree"))
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(img, M, (w, h))
    plt.imshow(rotated,cmap="gray")
    plt.title('rotated')
    plt.show()
```

```python
def crop():
    cropped = img[int(input("x1:")):int(input("x2:")), int(input("y1:")):int(input("y2:"))]
    plt.imshow(cropped,cmap="gray")
    plt.title('cropped')
    plt.show()
```

```python
def trans():
    (h, w) = img.shape[:2]
    x1,y1=input("give x and y coordinate for image translation").split(' ')
    M=np.float32([[1, 0,int(x1)],[0, 1,int(y1) ],[0, 0, 1]])
    translated = cv2.warpPerspective(img, M, (h, w))
    plt.imshow(translated,cmap="gray")
    plt.title('translated')
    plt.show()
```

```python
def shear():
    (h, w) = img.shape[:2]
    shx,shy=input("give shx and shy coordinate for image shearing").split(' ')
    M = np.float32([[1, float(shx), 0],[float(shy),1 , 0],[0, 0 , 1]])
    sheared = cv2.warpPerspective(img,M,(int(h*1.5),int(w*1.5)))
    plt.imshow(sheared,cmap="gray")
    plt.title('sheared')
    plt.show()
```

```python
def flip():
    flipped = cv2.flip(img,int(input('0: xaxis 1: yaxis -1:bothaxis')))
    plt.imshow(flipped,cmap="gray")
    plt.title('flipped')
    plt.show()    
```

```python
x=int(input('Select the operation you want to perform on input image.\n 1:resize, 2:rotate, 3:crop 4:translation 5:shearing, 6:flip'))
while(x!=0):
    if x==1: resize()
    elif x==2: rotate()
    elif x==3: crop()
    elif x==4: trans()
    elif x==5: shear()
    elif x==6:flip()
    else: print('Invalid Choice')
    x=int(input('Select the operation you want to perform on input image.\n 1:resize, 2:rotate, 3:crop 4:translation 5:shearing, 6:flip'))
```
