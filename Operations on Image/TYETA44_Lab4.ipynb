```python
#Write a Python Code for following Bitwise Operation:
#1. AND
#2. OR
#3. XOR
#4. NOT
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread('D:\TY ET A\CV\image1.jpg',0)  
image2 = cv2.imread('D:\TY ET A\CV\image2.jpg',0) 
print(image1.shape,image2.shape)
```

```python
plt.imshow(image1,cmap="gray")
plt.title('Original')
plt.show()
```

```python
plt.imshow(image2,cmap="gray")
plt.title('Original')
plt.show()
```

```python
def b_and():
    bitand=cv2.bitwise_and(image1, image2, mask = None)
    plt.imshow(bitand,cmap="gray")
    plt.title('AND')
    plt.show()    
```

```python
def b_or():
    bitor=cv2.bitwise_or(image1, image2, mask = None)
    plt.imshow(bitor,cmap="gray")
    plt.title('OR')
    plt.show()
```

```python
def b_xor():
    bitxor=cv2.bitwise_xor(image1, image2, mask = None)
    plt.imshow(bitxor,cmap="gray")
    plt.title('XOR')
    plt.show()
```

```python
def b_not():
    bitnot=cv2.bitwise_not(image1, mask = None)
    plt.imshow(bitnot,cmap="gray")
    plt.title('NOT')
    plt.show()
```

```python
x=int(input('Select the Transformation you want to perform on input image.\n 1:AND 2:OR 3:XOR 4:NOT : '))
while(x!=0):
    if x==1:b_and()
    elif x==2:b_or()
    elif x==3:b_xor()
    elif x==4:b_not()
    else: print('Invalid Choice')
    x=int(input('Select the Transformation you want to perform on input image.\n 1:AND 2:OR 3:XOR 4:NOT : '))
```
