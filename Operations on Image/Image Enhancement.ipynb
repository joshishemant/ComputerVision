```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread('D:\TY ET A\CV\sunflower.jpg',0)
plt.imshow(image1,cmap="gray")
plt.title('Original')
plt.show()
hist = cv2.calcHist(image1, [0], None, [256], [0, 256])
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

```python
def histeq():
    equ = cv2.equalizeHist(image1)
    plt.imshow(equ,cmap="gray")
    plt.title('Equalised')
    plt.show()
    hist = cv2.calcHist(equ, [0], None, [256], [0, 256])
    plt.plot(hist)
    plt.xlim([0, 256])
    plt.show()
```

```python
def contrast_stretch():
    def pixelVal(pix, r1, s1, r2, s2):
        if (0 <= pix and pix <= r1):
            return (s1 / r1)*pix
        elif (r1 < pix and pix <= r2):
            return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
        else:
            return ((255 - s2)/(255 - r2)) * (pix - r2) + s2 
    r1 = 70
    s1 = 0
    r2 = 140
    s2 = 255
    pixelVal_vec = np.vectorize(pixelVal)
    contrast_stretched = pixelVal_vec(image1, r1, s1, r2, s2)
    plt.imshow(contrast_stretched,cmap="gray")
    plt.title('contrast_stretch.jpg')
    plt.show()
```

```python
x=int(input('Select the operation you want to perform on input image.\n 1:Histogram equalization  2:Linear stretching: '))
while(x!=0):
    if x==1:histeq()
    elif x==2:contrast_stretch()
    else: print('Invalid Choice')
    x=int(input('Select the operation you want to perform on input image.\n 1:Histogram equalization  2:Linear stretching: '))
```
