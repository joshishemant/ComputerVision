```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread('D:\TY ET A\CV\image1.jpg',0)  
image2 = cv2.imread('D:\TY ET A\CV\image2.jpg',0) 
print(image1.shape,image2.shape)
```

```python
plt.imshow(image1,cmap="gray")
plt.title('Original')
plt.show()
```

```python
plt.imshow(image2,cmap="gray")
plt.title('Original')
plt.show()
```

```python
def substract():
    sub = cv2.subtract(image2, image1)
    plt.imshow(sub,cmap="gray")
    plt.title('Substracted')
    plt.show()
```

```python
def addition():
    add = cv2.add(image1, image2)
    plt.imshow(add,cmap="gray")
    plt.title('Added')
    plt.show()
```

```python
def multi():
    mul = cv2.multiply(image1, image2)
    plt.imshow(mul,cmap="gray")
    plt.title('Multiplied')
    plt.show()
```

```python
def div():
    div = cv2.divide(image1, image2)
    plt.imshow(div,cmap="gray")
    plt.title('Divided')
    plt.show()
```

```python
def b_and():
    bitand=cv2.bitwise_and(image1, image2, mask = None)
    cv2.imshow('bitand.jpg', bitand)
    cv2.waitKey(0)
    cv2.destroyAllWindows()    
```

```python
def b_or():
    bitor=cv2.bitwise_or(image1, image2, mask = None)
    cv2.imshow('bitor.jpg', bitor)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def b_xor():
    bitxor=cv2.bitwise_xor(image1, image2, mask = None)
    cv2.imshow('bitxor.jpg', bitxor)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def b_not():
    bitnot=cv2.bitwise_not(image1, mask = None)
    cv2.imshow('bitnot.jpg', bitnot)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
x=int(input('Select the Transformation you want to perform on input image.\n 1:Subs 2:Add 3:Multiply 4:Divide '))
while(x!=0):
    if x==1: substract()
    elif x==2: addition()
    elif x==3:multi()
    elif x==4:div()
    else: print('Invalid Choice')
    x=int(input('Select the Transformation you want to perform on input image.\n 1:Subs 2:Add 3:Multiply 4:Divide '))
```
